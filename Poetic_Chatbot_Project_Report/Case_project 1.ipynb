```python
import time
from google import genai
from google.colab import userdata
```

```python
# 1. Get Gemini API key from Colab's secure storage
key = userdata.get('API_key')
```

```python
#2. Typing animation function (character by character)
#    - Takes text and prints it slowly, like live typing
#    - 'delay' controls speed (default = 0.12 sec per character)
def stream(text, delay=0.12):
    print("\nBot:", end=" ", flush=True)  # Prefix for chatbot response
    for char in text:  # Loop through each character
        print(char, end="", flush=True)  # Print without newline
        time.sleep(delay)  # Small pause to simulate typing
    print("\n")  # Move to next line after finishing
```

```python
text='we are building a chatbot'
stream(text)
```

```python
# 3. Initialize Gemini client with API key
try:
 client = genai.Client(api_key=key)
except KeyError:
 print(f'Error with gemini api key. error message: {KeyError}')
# 4. Create a new chat session with Gemini
#    - Using the "gemini-2.0-flash" model (fast response model)
chat = client.chats.create(model="gemini-2.0-flash")
```

```python
# 5. Function to send user input to Gemini
#    - Wraps the input inside a poetic instruction
#    - Returns Gemini's text response
def ask_gemini(user_input):
 try:
  # Custom prompt to enforce poetic replies
  prompt = f"You are a poet chatbot. Respond to every question in a poetic way related to the topic, even if it's a serious question. Keep it short and poetic! Here's the question: {user_input}"
  # Send message to Gemini chat
  response = chat.send_message(prompt)
  # Extract text and remove extra spaces/newlines
  return response.text.strip()
 except Exception as e:
  return f"Oops, something went wrong: {e}"
```

```python
# 6. Main chatbot loop
#    - Keeps running until user types 'quit'
#    - Displays Gemini's response with typing effect
print("Welcome to your Gemini Chatbot! Type 'quit' to exit.")
while True:
 # Get input from user
 user_input = input("You: ")
 if user_input.lower() == 'quit':
  print("Bot: Bye for now!")
  break
 # Get Gemini's response
 response = ask_gemini(user_input)
 # Print response with typing animation
 stream(response)
```
