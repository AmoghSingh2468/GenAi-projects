```python
# Step 2: Import required libraries
import time  # For typing animation (delays)
from transformers import MarianMTModel, MarianTokenizer # Hugging Face tools for translation
```

```python
# Step 3: Load pre-trained English-to-French model
model_name = "Helsinki-NLP/opus-mt-en-fr"
tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)
```

```python
# Step 4: Typing animation function (character by character)
# This function makes the bot output look like it's "typing" naturally
def stream(text, delay=0.12):
    print("\nBot:", end=" ", flush=True)   # Print "Bot:" only once
    for char in text:                      # Loop through each character in the text
        print(char, end="", flush=True)    # Print characters one by one
        time.sleep(delay)                  # Small pause for typing effect
    print("\n")                            # Move to a new line after finishing
```

```python
# Step 5: Translation function
# Takes an English sentence, passes it through the model, and returns French translation
def translate_to_french(english_text):
    # Convert English text into model-readable tokens
    inputs = tokenizer(english_text, return_tensors="pt", padding=True)

    # Generate French translation tokens using the model
    translated_tokens = model.generate(**inputs)

    # Decode tokens back into French text
    french_translation = tokenizer.decode(translated_tokens[0], skip_special_tokens=True)

    return french_translation
```

```python
# Step 6: Chatbot loop
# Keeps running until the user types "quit"
print("Welcome to the English-to-French Translation Chatbot! Type 'quit' to exit.\n")

while True:
    user_input = input("You: ")     # Take input from user
    if user_input.lower() == "quit":   # Exit condition
        stream("Goodbye! Merci et à bientôt 👋")  # Typing effect farewell message
        break

    # Translate English → French
    french_text = translate_to_french(user_input)

    # Show translation with typing animation
    stream(french_text)
```
